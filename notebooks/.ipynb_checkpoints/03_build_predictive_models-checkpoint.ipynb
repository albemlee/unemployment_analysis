# Import Libraries

# Load Data Warehouse

```python
# Load birth and unemployment data into DataFrame (tidy_warehouse)
tidy_warehouse_df = pd.read_csv('../tidy_data/tidy_warehouse.csv')
tidy_warehouse_df.sort_values(by='year', inplace=True)
```

# Build Models

### Univariate Models

Naive Model

```python
# Extract years from 1968 - 2015
years = tidy_warehouse_df['year'].values[20:]

# Extract true average unemployment rate values from 1968 - 2015
true_values = tidy_warehouse_df['average_unemployment_rate'].values[20:]

# Build 1, 5, 10, and 20 year naive models
naive_1year_predictions = tidy_warehouse_df['average_unemployment_rate'].values[19:-1]
naive_5year_predictions = tidy_warehouse_df['average_unemployment_rate'].values[15:-5]
naive_10year_predictions = tidy_warehouse_df['average_unemployment_rate'].values[10:-10]
naive_20year_predictions = tidy_warehouse_df['average_unemployment_rate'].values[:-20]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=4,figsize=(18, 10))

# Plot predictions and actual values for each model
i = 0
for predictions in [naive_1year_predictions, naive_1year_predictions, naive_10year_predictions, naive_20year_predictions]:
    ax[0, i].plot(years, true_values)
    ax[0, i].plot(years, predictions)
    i += 1
    
# Plot error values for each model
i = 0
for predictions in [naive_1year_predictions, naive_1year_predictions, naive_10year_predictions, naive_20year_predictions]:
    ax[1, i].bar(years, true_values-predictions)
    i += 1

# Create legend for each plot
ax[0, 0].legend(['true unemployment', '1-year naive prediction'], loc='best')
ax[0, 1].legend(['true unemployment', '5-year naive prediction'], loc='best')
ax[0, 2].legend(['true unemployment', '10-year naive prediction'], loc='best')
ax[0, 3].legend(['true unemployment', '20-year naive prediction'], loc='best')
ax[1, 0].legend(['1-year naive error'], loc='best')
ax[1, 1].legend(['5-year naive error'], loc='best')
ax[1, 2].legend(['10-year naive error'], loc='best')
ax[1, 3].legend(['20-year naive error'], loc='best')
```

```python
# Calculate MSE for each model
naive_1year_mse = mean_squared_error(y_true=true_values, y_pred=naive_1year_predictions)
naive_5year_mse = mean_squared_error(y_true=true_values, y_pred=naive_5year_predictions)
naive_10year_mse = mean_squared_error(y_true=true_values, y_pred=naive_10year_predictions)
naive_20year_mse = mean_squared_error(y_true=true_values, y_pred=naive_20year_predictions)

# Print MSE values
print('1 year naive loss :', naive_1year_mse)
print('5 year naive loss :', naive_5year_mse)
print('10 year naive loss:', naive_10year_mse)
print('20 year naive loss:', naive_20year_mse)
```

Moving Average

```python
# Extract years from 1968 - 2015
years = tidy_warehouse_df['year'].values[20:]

# Extract true average unemployment rate values from 1968 - 2015
true_values = tidy_warehouse_df['average_unemployment_rate'].values[20:]

# Build 5, 10, and 20 year moving average (MA) models
ma_1year_predictions = naive_1year_predictions.copy()
ma_5year_predictions = tidy_warehouse_df['average_unemployment_rate'].rolling(window=5).mean()[19:-1]
ma_10year_predictions = tidy_warehouse_df['average_unemployment_rate'].rolling(window=10).mean()[19:-1]
ma_20year_predictions = tidy_warehouse_df['average_unemployment_rate'].rolling(window=20).mean()[19:-1]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=4,figsize=(18, 10))

# Plot predictions and actual values for each model
i = 0
for predictions in [ma_1year_predictions, ma_5year_predictions, ma_10year_predictions, ma_20year_predictions]:
    ax[0, i].plot(years, true_values)
    ax[0, i].plot(years, predictions)
    i += 1
    
# Plot error values for each model
i = 0
for predictions in [ma_1year_predictions, ma_5year_predictions, ma_10year_predictions, ma_20year_predictions]:
    ax[1, i].bar(years, true_values-predictions)
    i += 1

# Create legend for each plot
ax[0, 0].legend(['true unemployment', '1-year moving average prediction'], loc='best')
ax[0, 1].legend(['true unemployment', '5-year moving average prediction'], loc='best')
ax[0, 2].legend(['true unemployment', '10-year moving average prediction'], loc='best')
ax[0, 3].legend(['true unemployment', '20-year moving average prediction'], loc='best')
ax[1, 0].legend(['1-year moving average error'], loc='best')
ax[1, 1].legend(['5-year moving average error'], loc='best')
ax[1, 2].legend(['10-year moving average error'], loc='best')
ax[1, 3].legend(['20-year moving average error'], loc='best')
```

```python
# Calculate MSE for each model
ma_1year_mse = mean_squared_error(y_true=true_values, y_pred=ma_1year_predictions)
ma_5year_mse = mean_squared_error(y_true=true_values, y_pred=ma_5year_predictions)
ma_10year_mse = mean_squared_error(y_true=true_values, y_pred=ma_10year_predictions)
ma_20year_mse = mean_squared_error(y_true=true_values, y_pred=ma_20year_predictions)

# Print MSE values
print('1 year moving average loss :', ma_1year_mse)
print('5 year moving average loss :', ma_5year_mse)
print('10 year moving average loss:', ma_10year_mse)
print('20 year moving average loss:', ma_20year_mse)
```

```python
# Fit the model

# Separate label from features
X = input_df.drop(columns='average_unemployment_rate')
y = input_df['average_unemployment_rate']

# Separate into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Initialize models
xgb_base_model = XGBRegressor()
xgb_1year_model = XGBRegressor()
xgb_5year_model = XGBRegressor()
xgb_10year_model = XGBRegressor()
xgb_20year_model = XGBRegressor()
xgb_complete_model = XGBRegressor()

# Fit models
xgb_base_model.fit(X_train[base_features], y_train)
xgb_1year_model.fit(X_train[base_features + year_1_features], y_train)
xgb_5year_model.fit(X_train[base_features + year_5_features], y_train)
xgb_10year_model.fit(X_train[base_features + year_10_features], y_train)
xgb_20year_model.fit(X_train[base_features + year_20_features], y_train)
xgb_complete_model.fit(X_train.drop(columns='year'), y_train)

# Generate predictions
xgb_base_predictions = xgb_base_model.predict(X_test[base_features])
xgb_1year_predictions = xgb_1year_model.predict(X_test[base_features + year_1_features])
xgb_5year_predictions = xgb_5year_model.predict(X_test[base_features + year_5_features])
xgb_10year_predictions = xgb_10year_model.predict(X_test[base_features + year_10_features])
xgb_20year_predictions = xgb_20year_model.predict(X_test[base_features + year_20_features])
xgb_complete_predictions = xgb_complete_model.predict(X_test.drop(columns='year'))

# Calculate cross validation score
xgb_base_mse = mean_squared_error(y_true=y_test, y_pred=xgb_base_predictions)
xgb_1year_mse = mean_squared_error(y_true=y_test, y_pred=xgb_1year_predictions)
xgb_5year_mse = mean_squared_error(y_true=y_test, y_pred=xgb_5year_predictions)
xgb_10year_mse = mean_squared_error(y_true=y_test, y_pred=xgb_10year_predictions)
xgb_20year_mse = mean_squared_error(y_true=y_test, y_pred=xgb_20year_predictions)
xgb_complete_mse = mean_squared_error(y_true=y_test, y_pred=xgb_complete_predictions)

print('xgb base model loss:    ', xgb_base_mse)
print('xgb 1year model loss:   ', xgb_1year_mse)
print('xgb 5year model loss:   ', xgb_5year_mse)
print('xgb 10year model loss:  ', xgb_10year_mse)
print('xgb 20year model loss:  ', xgb_20year_mse)
print('xgb complete model loss:', xgb_complete_mse)
```
